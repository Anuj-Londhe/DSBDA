```python
sorted_rscore = sorted(ap['writing score'].fillna(0))
```

```python
# sorted(ap['writing score'])
print(sorted_rscore)
```

```python
# For outliers : Outlier is an observation in a given dataset that lies far from the
# rest of observations. Occurs due to variability in data.
q1 = np.percentile(sorted_rscore, 25)
q2 = np.percentile(sorted_rscore, 75)
print(q1, q2)
```

```python
# IQR: Inter Quartile Range: used to find outliers
IQR = q1-q2
lwr_bound = q1-(1.5*IQR)
upr_bound = q2-(1.5*IQR)
print(lwr_bound, upr_bound)
```

```python
r_outliers = []
for i in sorted_rscore:
    if i<lwr_bound or i>upr_bound:
        r_outliers.append(i)
print(r_outliers)
```
